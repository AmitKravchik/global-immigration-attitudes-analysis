```python
import requests

def fetch_articles_with_tone(query, tone_constraint, source_country, start_datetime, end_datetime, maxrecords=250, output_format="json"):
    """
    Fetch articles that match the given query and tone constraint from GDELT within a specified date range.
    
    Parameters:
        query (str): The base search query (e.g., "immigration").
        tone_constraint (str): The tone filter (e.g., "tone<-5" or "tone>5").
        start_datetime (str): Start date/time in YYYYMMDDHHMMSS format.
        end_datetime (str): End date/time in YYYYMMDDHHMMSS format.
        maxrecords (int): Maximum number of records to return.
        output_format (str): Format of the output, e.g., "json".
    
    Returns:
        JSON data containing the list of articles.
    """
    base_url = "https://api.gdeltproject.org/api/v2/doc/doc"
    
    # Combine the base query with the tone constraint
    full_query = f"{query} {tone_constraint} {theme} {source_country}"
    
    params = {
        "query": full_query,
        "mode": "artlist",
        "STARTDATETIME": start_datetime,
        "ENDDATETIME": end_datetime,
        "maxrecords": maxrecords,
        "format": output_format,
        "sort": "ToneDesc",
    }
    
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print("An error occurred:", e)
        return None

if __name__ == "__main__":
    # Example usage:
    # Search for articles about "immigration" with a negative tone (tone below -5)
    query = "immigration"
    tone_constraint = "tone>0 tone<0.999"
    theme = "IMMIGRATION"
    source_country = "sourcecountry:US"
    # Define a date range: January 1, 2024 to January 31, 2024
    start_datetime = "20250101000000"
    end_datetime = "20250101010000"
    
    articles_data = fetch_articles_with_tone(query, tone_constraint,source_country, start_datetime, end_datetime)
    if articles_data:
        print(articles_data)
```

```python
import requests
from bs4 import BeautifulSoup

def get_text_and_title(url):
    """
    Extracts the title and main body text of an article from a given URL.
    
    Parameters:
        url (str): The URL of the news article.
    
    Returns:
        A tuple (title, body) containing the article's title and main text.
    """
    try:
        # Send a GET request to the URL with a user-agent header to mimic a browser.
        headers = {'User-Agent': 'Mozilla/5.0 (compatible; YourScraper/1.0)'}
        response = requests.get(url, headers=headers, timeout=5)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None, None

    # Parse the HTML content.
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Extract title:
    # Most pages have the title in a <title> tag, but some may also have a <h1> within the article.
    title = None
    if soup.title:
        title = soup.title.get_text(strip=True)
    # Optionally, look for a primary headline element.
    h1 = soup.find('h1')
    if h1 and (not title or len(h1.get_text(strip=True)) > len(title)):
        title = h1.get_text(strip=True)
    
    # Extract article body:
    # Many news sites wrap the main text in an <article> tag.
    body = ""
    article_tag = soup.find('article')
    if article_tag:
        body = article_tag.get_text(separator=" ", strip=True)
    else:
        # If no <article> tag exists, try common container class names.
        for class_name in ['article-body', 'post-content', 'entry-content', 'story-body']:
            container = soup.find('div', class_=class_name)
            if container:
                body = container.get_text(separator=" ", strip=True)
                break
        # If still not found, use a fallback: the body tag itself.
        if not body:
            body_tag = soup.find('body')
            if body_tag:
                body = body_tag.get_text(separator=" ", strip=True)
    
    return body, title

```

```python
import requests

def fetch_timeline_tone_json(query, timespan="3months", timeline_smooth=5, output_format="json"):
    """
    Fetches a TimelineTone visualization from the GDELT DOC API in JSON format.
    
    Parameters:
        query (str): The search query (e.g., '"donald trump"').
        timespan (str): The time span to search (e.g., "3months", "1week").
        timeline_smooth (int): Smoothing window for the timeline (up to 30).
        output_format (str): Format of the output (e.g., "json").
        
    Returns:
        A JSON object containing the TimelineTone data.
    """
    base_url = "https://api.gdeltproject.org/api/v2/doc/doc"
    params = {
        "query": query,
        "mode": "timelineTone",  # Set mode to timelineTone
        "timespan": timespan,
        "TIMELINESMOOTH": timeline_smooth,
        "format": output_format
    }
    
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Check for HTTP errors
        return response.json()
    except requests.exceptions.RequestException as e:
        print("Error fetching data:", e)
        return None

if __name__ == "__main__":
    # Example usage:
    # Retrieve timeline tone data for articles mentioning "donald trump" over the last 3 months,
    # with a smoothing window of 5 time steps.
    query = '"immigration"'
    timeline_tone_data = fetch_timeline_tone_json(query)
    if timeline_tone_data:
        print(timeline_tone_data)
```

```python
import requests

def fetch_wordcloud_json(query, maxrecords=100, output_format="json"):
    """
    Fetches a word cloud (based on image tags) from the GDELT DOC API in JSON format.
    
    Parameters:
        query (str): The search query (e.g., "immigration").
        maxrecords (int): Maximum number of records to include (default is 100).
        output_format (str): The desired output format ("json" in this example).
        
    Returns:
        A JSON object containing the word cloud data.
    """
    base_url = "https://api.gdeltproject.org/api/v2/doc/doc"
    
    # Use the "wordcloudimagetags" mode as described in the documentation.
    params = {
        "query": query,
        "mode": "WordCloudImageWebTags",  
        "maxrecords": maxrecords,
        "format": output_format
    }
    
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raises an HTTPError if the response status is not OK.
        return response.json()
    except requests.exceptions.RequestException as e:
        print("Error fetching word cloud data:", e)
        return None

if __name__ == "__main__":
    # Example usage: Retrieve a word cloud for articles mentioning "immigration"
    query = '"trump"'
    wordcloud_data = fetch_wordcloud_json(query)
    if wordcloud_data:
        print(wordcloud_data)
```

```python
import requests

def fetch_wordcloud_json(query, maxrecords=100, output_format="json", timespan="3months"):
    """
    Fetches a word cloud (based on image tags) from the GDELT DOC API in JSON format.
    
    Parameters:
        query (str): The base search query.
        maxrecords (int): Maximum number of records.
        output_format (str): Desired output format.
        timespan (str): Timespan for the search (e.g., "3months").
        
    Returns:
        JSON object containing the word cloud data.
    """
    base_url = "https://api.gdeltproject.org/api/v2/doc/doc"
    
    # Adjust the query to force image-tagged articles; note the use of AND and quotes.
    # You can change "immigration" inside imagewebtag to any other keyword if desired.
    full_query = f'{query} AND imagewebtag:"immigration"'
    
    params = {
        "query": full_query,
        "mode": "WordCloudImageTags",  # Try capitalized mode as per documentation
        "maxrecords": maxrecords,
        "timespan": timespan,
        "format": output_format
    }
    
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raises an error for non-200 responses
        json_data = response.json()
        # Debug: print the raw JSON to inspect its structure
        print("Raw JSON response:", json_data)
        return json_data
    except requests.exceptions.RequestException as e:
        print("Error fetching word cloud data:", e)
        return None

if __name__ == "__main__":
    # Example usage: Search for "immigration" with an image constraint.
    query = "immigration"
    wordcloud_data = fetch_wordcloud_json(query)
    if wordcloud_data:
        print(wordcloud_data)
```
